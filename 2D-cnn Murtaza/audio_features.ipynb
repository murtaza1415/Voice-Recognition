### MFCC: (Old - makes less data)

```python
#Load the data
audio_rate = 16000
n_mfcc = 60
no_seconds = 4
speakers = glob("data/LibriSpeech-100/train-clean-100/*/")

x_and_y = []
x_train = []
y_train = []
no_speakers_loaded = 0

for speaker in speakers:
    label = os.path.split(os.path.split(speaker)[0])[1]                  #Read speaker ID from path.
    audio_files = glob(speaker + '*.flac')
    for audio_file in audio_files:
        audio_data, _ = librosa.load(sr=audio_rate, mono=True, path=audio_file)
        assert _ == audio_rate        
        assert len(audio_data) >= (audio_rate*no_seconds)
        audio_data = audio_data[0:audio_rate*no_seconds]                                         #4 sec crop.
        audio_mfcc = librosa.feature.mfcc(y=audio_data, sr=audio_rate,  n_mfcc=n_mfcc)
        audio_mfcc = np.reshape(audio_mfcc, (-1,audio_mfcc.shape[1],1))
        x_and_y.append([audio_mfcc, label])
    no_speakers_loaded += 1
    print(no_speakers_loaded)
```

```python
#Correctly reorganize arrays
random.shuffle(x_and_y)        
x_and_y = np.asarray(x_and_y)
x_train = x_and_y[:,0]
x_train = np.stack(x_train)
y_train = x_and_y[:,1]
y_train = np.reshape(y_train, (-1,1))
```

```python
#Save the numpy data
np.save('data/x_mfcc.npy', x_train)
np.save('data/y_mfcc.npy', y_train)
```

### MFCC: (with Normalization)

```python
#Load the data
audio_rate = 16000
n_mfcc = 60
no_seconds = 4
speakers = glob("data/LibriSpeech-100/train-clean-100/*/")

x_and_y = []
x_train = []
y_train = []
no_speakers_loaded = 0

for speaker in speakers:
    label = os.path.split(os.path.split(speaker)[0])[1]                  #Read speaker ID from path.
    audio_files = glob(speaker + '*.flac')
    for audio_file in audio_files:
        audio_data, _ = librosa.load(sr=audio_rate, mono=True, path=audio_file)
        assert _ == audio_rate
        
        no_clips = len(audio_data) / (audio_rate*no_seconds)
        #print(no_clips)
        for i in range(0, no_clips):                                                                #4 sec clips
            audio_clip = audio_data[(audio_rate*no_seconds)*i:(audio_rate*no_seconds)*(i+1)]
            audio_clip = audio_clip/abs(audio_clip).max()                                           #Normalization.
            audio_mfcc = librosa.feature.mfcc(y=audio_clip, sr=audio_rate,  n_mfcc=n_mfcc)
            audio_mfcc = np.reshape(audio_mfcc, (-1,audio_mfcc.shape[1],1))
            x_and_y.append([audio_mfcc, label])                                        
    no_speakers_loaded += 1
    print(no_speakers_loaded)
```

```python
#Correctly reorganize arrays
random.shuffle(x_and_y)        
x_and_y = np.asarray(x_and_y)
x_train = x_and_y[:,0]
x_train = np.stack(x_train)
y_train = x_and_y[:,1]
y_train = np.reshape(y_train, (-1,1))
```

```python
#Save the numpy data
np.save('data/LibriSpe/x_.npy', x_train)
np.save('data/LibriS/y_.npy', y_train)
```

```python
#Load the data
audio_rate = 16000
n_mfcc = 60
no_seconds = 4
speakers = glob("data/LibriSpeech-100/train-clean-100/*/")

x_and_y = []
x_train = []
y_train = []
no_speakers_loaded = 0

for speaker in speakers:
    label = os.path.split(os.path.split(speaker)[0])[1]                  #Read speaker ID from path.
    audio_files = glob(speaker + '*.flac')
    for audio_file in audio_files:
        audio_data, _ = librosa.load(sr=audio_rate, mono=True, path=audio_file)
        assert _ == audio_rate
        
        no_clips = len(audio_data) / (audio_rate*no_seconds)
        #print(no_clips)
        for i in range(0, no_clips):                                                                #4 sec clips
            audio_clip = audio_data[(audio_rate*no_seconds)*i:(audio_rate*no_seconds)*(i+1)]
            audio_clip = noise.insert_noise(audio_clip, audio_rate, no_seconds)                     #Insert noise
            audio_mfcc = librosa.feature.mfcc(y=audio_clip, sr=audio_rate,  n_mfcc=n_mfcc)
            audio_mfcc = np.reshape(audio_mfcc, (-1,audio_mfcc.shape[1],1))
            x_and_y.append([audio_mfcc, label])                                        
    no_speakers_loaded += 1
    print(no_speakers_loaded)
```
